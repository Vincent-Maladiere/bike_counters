```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fr-en-calendrier-scolaire.csv", sep=";")
df.head()
```

Paris belongs to the Zone C

```python
data = df[df["Académies"]=="Paris"]
data.head()
```

```python
df['Date de début'] = pd.to_datetime(df['Date de début'])
df['Date de fin'] = pd.to_datetime(df['Date de fin'])

# Fonction pour générer une liste de dates entre deux dates
def generate_date_range(start_date, end_date):
    return pd.date_range(start=start_date.date(), end=end_date.date()).tolist()

# Appliquer la fonction à chaque ligne du DataFrame
all_dates = []
for index, row in df.iterrows():
    dates = generate_date_range(row['Date de début'], row['Date de fin'])
    all_dates.extend(dates)

# Convertir la liste de dates en DataFrame et en datetime
vacation_dates_df = pd.DataFrame(all_dates, columns=['date'])
vacation_dates_df['date'] = pd.to_datetime(vacation_dates_df['date'])

vacation_dates_df.head()
```

```python
# Convertir la date que vous vérifiez en objet datetime
check_date = pd.to_datetime('2017-12-24')

# Vérifier si la date est présente dans le DataFrame
print(check_date in vacation_dates_df['date'].values)
```

```python
vacation_dates_df.to_csv("Propre_vacances_scolaires.csv", index=False)
```
