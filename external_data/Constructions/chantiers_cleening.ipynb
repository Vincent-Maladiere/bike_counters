```python
import pandas as pd
```

```python
df = pd.read_csv('chantiers-perturbants.csv', sep=";")
df.head()
```

```python
# Convertir la colonne 'Date et heure de comptage' en type datetime
df['Date de début'] = pd.to_datetime(df['Date de début'])

# Classer les données par ordre chronologique
df = df.sort_values(by='Date de début', ascending=True)

df.head()
```

```python
df.columns
```

```python
df = df.drop(columns= ['Identifiant', 'Identifiant CTV', "Code postal de l'arrondissement",
       'Numéro de STV', 'Typologie', "Maitre d'ouvrage", 'Objet', "Voie(s)",
       'Description', 'Précisions de localisation', 
       'URL LettreInfoChantier', 'geo_shape', "Détail de l'impact sur la circulation", 
       "Impact sur la circulation"])

df = df[df['Statut'].isin([2,4])] # garde que les travaux actifs sur la période mentionnée

df.head()
```

```python
df = df.drop(columns= ["Niveau de perturbation", "Statut"])
df.head()
```

```python
df.to_csv("Propres_chantiers_GPS.csv", index=False)
```
