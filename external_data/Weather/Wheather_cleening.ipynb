```python
colonnes_a_conserver = [
    'Date et heure', 'Température de l\'air', 'Humidité relative',
    'Visibilité horizontale', 'Vitesse du vent', 'Rafale de vent sur 10 minutes',
    'Précipitations sur 3 heures'
]

df = df[colonnes_a_conserver]
df['Date et heure'] = pd.to_datetime(df['Date et heure'])

df.head()
```

On a un relevé météorologique toutes les 3 heures alors que notre dataset a un relevé toutes les heures. Il faut donc combler le gap. 

```python
# Créer un DataFrame avec des intervalles horaires
date_min = df['Date et heure'].min()
date_max = df['Date et heure'].max()
date_range = pd.date_range(start=date_min, end=date_max, freq='h')
df_hourly = pd.DataFrame(date_range, columns=['Date et heure'])

# Fusionner les données horaires avec les données originales
df_hourly = df_hourly.merge(df, on='Date et heure', how='left')

df_hourly.head()
```

```python
# Remplir les valeurs manquantes pour les autres colonnes
for col in colonnes_a_conserver[1:]:
    df_hourly[col] = df_hourly[col].ffill()

df_hourly.head()
```

```python
# Calculer les précipitations horaires en utilisant la moyenne des précipitations sur les 3 heures précédentes
df_hourly['Précipitations horaires'] = df_hourly['Précipitations sur 3 heures'] / 3

df_hourly.head()
```

```python
df_hourly = df_hourly.drop(columns="Précipitations sur 3 heures")
df_hourly.head()
```

```python
df_hourly.to_csv("Propre_meteo.csv", index=False)
```
