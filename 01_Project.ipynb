```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skrub import TableReport
```

```python
threshold = 0.5 * len(external_data)
external_data = external_data.dropna(thresh=threshold, axis=1)
```

```python
external_data.isna().sum()
```

```python
num_cols = external_data.select_dtypes(include=[np.number]).columns
for col in num_cols:
    if external_data[col].isna().sum() > 0:
        external_data.loc[:, col] = external_data[col].fillna(external_data[col].median())

categ_cols = external_data.select_dtypes(include=["object"]).columns
for col in categ_cols:
    if external_data[col].isna().sum() > 0:
        external_data.loc[:, col] = external_data[col].fillna(external_data[col].mode()[0])
```

```python
external_data.isna().sum()
```

```python
external_data["date"] = pd.to_datetime(external_data["date"])
```

```python
train_df = pd.read_parquet("/Users/marcosalerno/Downloads/msdb-2024/train.parquet", engine="pyarrow")
```

```python
train_df.dtypes
```

```python
train_df.isna().sum()
```

```python
train_df["hour"] = train_df["date"].dt.hour
train_df["day_of_week"] = train_df["date"].dt.dayofweek
train_df["is_weekend"] = (train_df["day_of_week"] >= 5).astype(int)
```

```python
print("Time range of train_df:", train_df['date'].min(), "-", train_df['date'].max())
print("Time range of external_data:", external_data['date'].min(), "-", external_data['date'].max())
```

```python
plt.figure(figsize=(10, 6))
train_df['log_bike_count'].hist(bins=30)
plt.title('log_bike_count distribution')
plt.xlabel('log_bike_count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=train_df["log_bike_count"], color="skyblue")
plt.title("Boxplot of log_bike_count")
plt.xlabel("log_bike_count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
train_df.groupby("hour")["log_bike_count"].mean().plot(kind="bar")
plt.title("Average log_bike_count by hour of the day")
plt.xlabel("Hour of the day")
plt.ylabel("Average log_bike_count")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
train_df.groupby("day_of_week")["log_bike_count"].mean().plot(kind="bar")
plt.title("Average log_bike_count by day of the week")
plt.xlabel("Day of the week") # (0=Monday, 6=Sunday)
plt.ylabel("Average log_bike_count")
plt.xticks(rotation=0)
plt.show()
```
